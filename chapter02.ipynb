```python
# Combing data (รูปที่ 2.25)
import pandas as pd 
df_products = pd.DataFrame({
    'Year':pd.Series([2017, 2018, 2019, 2020, 2021, 2022, 2023]),
    'Price':pd.Series([199, 199, 199, 299, 299, 349, 349]),
    'Version':pd.Series(['v1', 'v1', 'v1', 'v2', 'v2', 'v2', 'v3'])})
df_products          
```

```python
# (รูปที่ 2.25) (ต่อ)
df_revenue = pd.DataFrame({
    'Year':pd.Series([2018, 2019, 2020, 2021, 2022, 2023]),
    'Revenue':pd.Series([9473, 8422, 9987, 7994, 9530, 9444])})
df_revenue
```

```python
# Combing data (รูปที่ 2.25)
df = pd.merge(df_products, df_revenue)
df
```

```python
# Combing data ระบุคอลัมน์ (รูปที่ 2.26)
df_Year = pd.merge(df_products, df_revenue, on='Year')
df_Year
```

```python
# join แบบ outer (รูปที่ 2.28) 
df_outer = pd.merge(df_products, df_revenue, how='outer')
df_outer
```

```python
# ตัวอย่าง 2.3 (รูปที่ 2.34)
import pandas as pd 
df1 = pd.read_csv('data/timeSpent.csv')
df2 = pd.read_csv('data/cost.csv')
df1.head()
df2.head()

df = df1.merge(df2, on='users', how='left')
df.head()

df = df.fillna(0)
df.head()

df.describe()
```

```python
# กิจกรรม 2.1  (รูปที่ 2.50)
import pandas as pd
import copy

sales = pd.read_csv('data/sales.csv')
sales.head()
sales.dtypes
sales.info()

# รวมคอลัมน์ 'Product' + 'line'
sales['Product line'] = sales['Product']+' '+sales['line']
sales = sales.drop(['Product', 'line'], axis=1)
sales.head()

sales.groupby(['Product.1','type'])['Year'].count()

# รวมคอลัมน์ 'Product.1' + 'type'
sales['Product type'] = sales['Product.1']+' '+sales['type']
sales = sales.drop(['Product.1', 'type'], axis=1)
sales.head()

# รวมคอลัมน์ 'Product.2' + 'Order' + 'method'
sales['Product'] = sales['Product.2']+' '+sales['Order']\
                   +' '+sales['method']
sales = sales.drop(['Product.2', 'Order', 'method'], axis=1)
sales.head()

sales.groupby(['type.1'])['Year'].count()

sales = sales.rename(columns = {'type.1':'Order method'})
sales.head()

sales = sales.fillna('')

sales['Retailer country'] = sales['Retailer']+' '+sales['country']
sales = sales.drop(['Retailer', 'country'], axis=1)
sales.head()

sales.to_csv('sales_edit.csv')
```
